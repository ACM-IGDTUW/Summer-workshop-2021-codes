## List comprehensions

````
# https://www.geeksforgeeks.org/comprehensions-in-python/ 

# looping basics


input_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]
  
output_list = [] 
  
# Using loop for constructing output list 
i = 0
for var in input_list: 
    if i%2 == 0:
        continue
    if var % 2 == 0: 
        output_list.append(var) 
    i+=1

  
print("Output List using for loop:", output_list) 
````

````
input_list = [1, 2, 3, 4, 5, 6, 7, 8, 9] 

list_using_comp = [(var) if (var%2 == 0) else (33 if (var==3) else 0) for var in input_list] 

print("Output List using list comprehensions:", list_using_comp) 

def class_work():
    # do list comprehension when you want to put 33 in place of 3
    # for the rest of the odd-numbers put 0
    # so that the output is
    # [0, 2, 33, 4, 0, 6, 0, 8, 0]
    # send me the one line code in chat
    return
````

```
You can use list comprehension to replace many for and while blocks.
List comprehension is faster because it is optimized for the Python

Besides the syntactic benefit of list comprehensions, they are often
as fast or faster than equivalent use of map. [4]
```

## Timing

````
import timeit

def lc():
    ml = [x for x in range(5000000) if x%2==0] #50 lacs
    return ml

def nlc():
    ml = []
    for x in range(5000000):
        if x%2==0:
            ml.append(x)
    return ml

print(timeit.timeit(stmt=lc, number=10))
print(timeit.timeit(stmt=nlc, number=10))
````

````
import time

start_time = time.time() # seconds from 1970 till now
a = lc()
end_time = time.time()
print(end_time-start_time) #seconds from 1970 till now + seconds of lc() - seconds from 1970 till now

start_time = time.time()
a = nlc()
print(time.time()-start_time)
````

````
def set_method(dost,dushman,dushman_ke_dushman):
    s1 = set(dost)
    s2 = set(dushman) 
    s3 = set(dushman_ke_dushman)
    return list( s1.union(s3.difference(s2)) ) # s1 - (s3 - s2)

print(set_method(dost, dushman, dushman_ke_dushman))

# why not ordered? because hashing [5]
````
