Passing values with get request 

```python
payload = {'status': '1323'}
r = requests.get('https://api.twitter.com/1.1/statuses/update.json', params=payload)
```

```python
r.status_code
```

```python
r.json()
```

### POST requests

Opposite of get request. Used to sedn data to a server, but it can recive data too in response.

```python
payload = {'name1': 'admin', 'name2': 'password'}
r = requests.post("http://httpbin.org/post", data=payload)
```

```python
r.url
```

```python
r.json()
```

Send images using POST request

![alt text](img1.jpg)  ![alt text](img2.jpg)

```python
url = 'http://httpbin.org/post'
file_list = [
    ('image', ('img1.jpg', open('img1.jpg', 'rb'), 'image/jpg')),
    ('image', ('img2.jpg', open('img2.jpg', 'rb'), 'image/jpg'))
]

r = requests.post(url, files=file_list)
```

```python
r.status_code
```

```python
r.text
```

Other Request methods are PUT, DELETE, HEAD, OPTION which you can read [here](https://developer.mozilla.org/en-US/docs/Web/HTTP/Methods) about.

Handling Time outs

```python
requests.get('http://www.google.com', timeout=1)
```

### Custom headers

```python
custom_header = {'user-agent': 'customUserAgent'}
r = requests.get('https://httpbin.org', headers=custom_header)
```

```python
r.status_code
```
